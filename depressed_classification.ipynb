```python
import numpy as np
colu=["Age","Number_children","education_level","total_members","gained_asset","durable_asset","save_asset","living_expenses","other_expenses","incoming_agricultural","farm_expenses","lasting_investment","no_lasting_investmen","Ville_id"]
def get_outliers_count(df):
    outliers_count = {}
    
    for col in colu:
        q1, q3 = df[col].quantile([0.25, 0.75])
        iqr = q3 - q1
        lower_fence, upper_fence = q1 - 1.5 * iqr, q3 + 1.5 * iqr
        outliers_count[col] = ((df[col] < lower_fence) | (df[col] > upper_fence)).sum()
    
    return outliers_count

def impute_outliers_with_median(df, outliers_count_dict):
    for col, count in outliers_count_dict.items():
        if count > 2:
            q1, q3 = df[col].quantile([0.25, 0.75])
            iqr = q3 - q1
            lower, upper = q1 - 1.5 * iqr, q3 + 1.5 * iqr
            df[col] = np.where(df[col].between(lower, upper), df[col], df[col].median())

    return df
get_outliers_count(d)
d=impute_outliers_with_median(d,get_outliers_count(d))
```

```python
d1=d
x = d1.drop(["depressed"], axis=1) 
y = d1["depressed"]
y
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2, random_state=50)
xtrain
```

```python
from imblearn.over_sampling import SMOTE 
sm = SMOTE(k_neighbors=2) 
X_train_res, y_train_res = sm.fit_resample(xtrain, ytrain) 
```

```python

colu=["Age","Number_children","education_level","total_members","gained_asset","durable_asset","save_asset","living_expenses","other_expenses","incoming_agricultural","farm_expenses","lasting_investment","no_lasting_investmen","Ville_id"]
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train_res[colu]=sc.fit_transform(X_train_res[colu])
X_train_res[colu]
X_train_res
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix

models = {
    "Logistic Regression": LogisticRegression(),
    "Random Forest": RandomForestClassifier(),
    "Support Vector Machine": SVC(),
    "DecisionTree":DecisionTreeClassifier(),
    "KNearestNeighbors":KNeighborsClassifier()}
for name, model in models.items():
    model.fit(X_train_res, y_train_res)
    ypred=model.predict(xtest)
    score=model.score(xtest,ytest)
    cn=confusion_matrix(ytest,ypred)
    print(model,score)
```

```python
model.score(X_train_res, y_train_res)
```

```python
from sklearn.model_selection import GridSearchCV
rf = RandomForestClassifier(random_state=42)
param_grid = {
    "n_estimators": [50, 100, 200, 300],    # Number of trees in the forest
    "max_depth": [None, 10, 20, 30],       # Maximum depth of trees
    "min_samples_split": [2, 5, 10],       # Min samples required to split a node
    "min_samples_leaf": [1, 2, 4],         # Min samples required at leaf node
    "bootstrap": [True, False]             # Whether bootstrap sampling is used
}
grid_search = GridSearchCV(
    estimator=rf, 
    param_grid=param_grid, 
    scoring="accuracy",  # Metric to evaluate
    cv=5,                # 5-fold cross-validation
    verbose=2,           # Show progress
    n_jobs=-1            # Use all available CPU cores
)


grid_search.fit(X_train_res, y_train_res)
best_model = grid_search.best_estimator_
test_score = best_model.score(xtest, ytest)
print("Test Accuracy:", test_score)
```
